```python
# text file with car Num as TS XX A-Z XXXX , # of seats , # of bags
import random
import string
file=open("DataBase.txt","w+")
file.write('CarNum,numofseats,numofbags,type,When\n')
typ=['SUV','Hatchback','Sedan']
whn=['Now','Later']
for i in range(101):
    
    file.write('TS '+str(random.randint(1,100))+" "+ random.choice(string.ascii_uppercase) + random.choice(string.ascii_uppercase) +" "+ str(random.randint(1,10))+ str(random.randint(1,10))+ str(random.randint(1,10))+ str(random.randint(1,10))+","+str(random.randint(2,8))+","+str(random.randint(3,8))+","+random.choice(typ)+","+random.choice(whn)+'\n')
    
file.close()
```
